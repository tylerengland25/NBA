```python
# Convert 'date' column to Date object
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df.info()
```

```python
total_df = df[[col for col in df.columns if col.split('_')[-1] == 'total'or col == 'date' or col == 'year' or col == 'month']]
q1_df = df[[col for col in df.columns if col.split('_')[-1] == 'q1'or col == 'date' or col == 'year' or col == 'month']]
q2_df = df[[col for col in df.columns if col.split('_')[-1] == 'q1'or col == 'date' or col == 'year' or col == 'month']]
q3_df = df[[col for col in df.columns if col.split('_')[-1] == 'q1'or col == 'date' or col == 'year' or col == 'month']]
q4_df = df[[col for col in df.columns if col.split('_')[-1] == 'q1'or col == 'date' or col == 'year' or col == 'month']]
```

```python
# Create target varibale
df['3p_total'] = df['3p_visitor_total'] + df['3p_home_total']

# Change plot size
fig = plt.gcf()
fig.set_size_inches(12, 7)

# Plot lineplot of Total 3pt Made over Time
chart = sns.lineplot(x='year', y='3p_total', data=df).set(title='3pt Made vs Time')
plt.show()
```

```python
# Change plot size
fig = plt.gcf()
fig.set_size_inches(12, 7)

# Plot boxplot of Total 3pt Made over Time
chart = sns.boxplot(x='year', y='3p_total', data=df).set(title= '3pt Made vs Time')
plt.xticks(rotation=45)
plt.show()
```
